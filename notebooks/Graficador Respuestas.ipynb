# Graficador Respuesta

## Respuesta Normalizada Dataset 1

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy

t = sympy.symbols('t', positive=True)
vh = 0.0035
```

```python
a1 = -93.3393
w1 = 359.9558

f1 = lambda t: -vh * np.exp(a1*t) * (np.cos(w1*t) - (a1/w1) * np.sin(w1*t))+ vh 

x = np.linspace(0,55,1000)
    
plt.plot(x, f1(x/1000)*1000, 'r--')
plt.grid()
# show the plot
plt.savefig('/tmp/demo1.png', transparent=True, dpi=500)
plt.show()
```

## Respuesta Normalizada Dataset 2

```python
r1 = -85.11
r2 = -101.01 

f1 = lambda t: vh * (-6.349 * np.exp(r1*t)+5.35*np.exp(r2*t) +1)

x = np.linspace(0,90,1000)
    
plt.plot(x, f1(x/1000)*1000, 'r--')
plt.grid()
# show the plot
plt.savefig('/tmp/demo2.png', transparent=True, dpi=500)
plt.show()
```
