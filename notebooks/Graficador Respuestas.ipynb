# Graficador Respuesta

## Respuesta Normalizada Dataset 1

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy

t = sympy.symbols('t', positive=True)
vh = 0.0035
```

```python
a1 = -93.3393
w1 = 359.9558

f1 = lambda t: -vh * np.exp(a1*t) * (np.cos(w1*t)+ (a1/w1) * np.sin(w1*t))+ vh 

x = np.linspace(0,55,1000)
    
plt.plot(x, f1(x/1000)*1000, 'r--')
plt.grid()
# show the plot
plt.savefig('/tmp/demo.png', transparent=True, dpi=500)
plt.show()
```
